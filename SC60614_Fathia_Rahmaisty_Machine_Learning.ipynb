## **3.3 Periksa Distribusi Data & Outlier**

```python
# Import library visualisasi data
import plotly.express as px

# Ambil kolom numerik saja pada data
number_data = data_heart.select_dtypes(include = 'number')

# Definisikan warna
colors = ["#800000", "#006994", "#228B22", "#FFC0CB", "#BA55D3", "#FFA500", "#808080"]

for col, colour in zip(number_data.columns, colors):
    # Buat horizontal box plot
    fig = px.histogram(
        data_heart[col],
        color_discrete_sequence  = [colour],
        marginal = 'box',
        nbins = 50
    )

    fig.update_traces(
        marker_line_width = 1,
        marker_line_color = 'white'
    )

    # Update layout
    fig.update_layout(
        width = 1200,
        height = 350,
        showlegend = False,
        plot_bgcolor = 'rgba(0, 0, 0, 0)',
        title = f"<b>Box Plot <span style='color:{colour}'>{col}</span></b>",
        yaxis = dict(
            title = '',
            showgrid = False,
            showline = False,
            showticklabels = False,
            zeroline = False,
        ),
        xaxis = dict(
            title = col,
            showgrid = False,
            showline = True,
            showticklabels = True,
            zeroline = False,
        )
    )

    # Tampilkan grafik
    fig.show()
```

```python
import pandas as pd
import numpy as np

def remove_outliers_zscore(df_without_duplicates, column_name, threshold=3):
  mean = df_without_duplicates[column_name].mean()
  std = df_without_duplicates[column_name].std()
  abs_z_scores = np.abs(df_without_duplicates[column_name] - mean) / std
  return df_without_duplicates.loc[abs_z_scores < threshold]

# Example usage with threshold of 3 standard deviations
df_filtered1 = remove_outliers_zscore(df_without_duplicates.copy(), 'chol')
df_filtered2 = remove_outliers_zscore(df_filtered1.copy(), 'trestbps')
df_final = remove_outliers_zscore(df_filtered1.copy(), 'thalach')
print(df_final)
```

# **4. Data Preprocessing**

## **4.1 One Hot Encoding**

```python
# categorical columns
cp = pd.get_dummies(data_heart['cp'],prefix= 'cp')
thal = pd.get_dummies(data_heart['thal'],prefix = 'thal')
slope = pd.get_dummies(data_heart['slope'],prefix = 'slope')

#merge dummy variables to main data frame
frames = [ data_heart,cp,thal,slope]
data_heart = pd.concat(frames,axis=1)

#Dropping the unnecessory variables
data_heart=data_heart.drop(columns = ['cp','thal','slope'])
data_heart.head()
```

```python
#Missing values
data_heart.isnull().sum()
```

## **4.2 Separating Target Variable**

```python
#Declaring Dependent and Independent variables
X = data_heart.drop(['target'],axis=1)
y = data_heart['target']
```

## **4.3 Data Normalization**

```python
from sklearn.preprocessing import MinMaxScaler
X = MinMaxScaler().fit_transform(X)
```

## **4.4 Splitting Training and Testing Dataset**

```python
#Splitting x and y into training and testing datasets
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2,random_state=4)
```

Karena PyCaret adalah library low-code machine learning yang mudah digunakan dan menghabiskan lebih sedikit waktu karena hanya menggunakan sedikit kode, PyCaret akan membantu memilihkan model yang tepat untuk prediksi.

## **7.2 Akses Data**

```python
import pandas as pd
# Memuat dataset
data = pd.read_csv('heart.csv')
```

## **7.3 Membandingkan (Compare) Model**

```python
from pycaret.classification import *
# Inisialisasi setup PyCaret
setup(data=data, target='target')

# Membandingkan dan memilih model terbaik
best_model = compare_models()
```

```python
# Tampilkan best model
display(best_model)
```

## **7.4 Tuning Model**

```python
# Melakukan tuning terhadap model terbaik
tuned_model = tune_model(best_model)
```

```python

# Membuat model Extra Trees Classifier
extratree_model = create_model('et')

# Melakukan tuning hyperparameter
tuned_extratree_model = tune_model(extratree_model)

# Evaluasi model
evaluate_model(tuned_extratree_model)
```

## **7.5 Analisis Model**

```python
plot_model(estimator = tuned_extratree_model, plot = 'auc')
```

AUC mengukur sejauh mana model mampu membedakan antara kelas positif dan negatif. AUC berkisar antara 0 hingga 1. Pada kurva terlihat bahwa Nilai AUC sebesar 1 menunjukkan model memiliki akurasi yang cukup tinggi.

```python
plot_model(estimator = tuned_extratree_model, plot = 'learning')
```


Dari hasil Learning Curve di atas terlihat bahwa training score lebih tinggi daripada cross-validation score. Hal ini menunjukkan adanya potensi overfitting pada model.

```python
plot_model(estimator = tuned_extratree_model, plot = 'confusion_matrix')
```

Pada Confusion Matrix, dapat dilihat bahwa

*   Terdapat 6 false negatives, yaitu sampel yang sebenarnya memiliki heart disease (kelas 1), tetapi model memprediksi sebagai tidak memiliki heart disease (kelas 0).
* Terdapat 152 true positives, yaitu sampel yang sebenarnya memiliki heart disease (kelas 1) dan model memprediksi dengan benar sebagai memiliki heart disease (kelas 1).
* Terdapat 150 true negatives, yaitu sampel yang sebenarnya tidak memiliki heart disease (kelas 0) dan model memprediksi dengan benar sebagai tidak memiliki heart disease (kelas 0).
* Terdapat 0 false positives, yaitu sampel yang sebenarnya tidak memiliki heart disease (kelas 0), tetapi model memprediksi sebagai memiliki heart disease (kelas 1).

## **Evaluasi Kinerja Model**

```python
# Membuat prediksi menggunakan model yang telah dituning
predictions = predict_model(tuned_model, data=data)

# Menampilkan evaluasi hasil prediksi
display(predictions)
```

1. Nilai Accuracy 1 menunjukkan bahwa model mengklasifikasikan semua sampel dengan benar. Model memberikan prediksi yang tepat untuk seluruh dataset.

2. Nilai AUC 1 menunjukkan bahwa model memiliki kemampuan yang sangat baik dalam membedakan kelas positif dan kelas negatif.

3. Nilai Recall 1 menunjukkan bahwa model mampu mengidentifikasi semua sampel positif yang ada. Tidak ada satupun sampel positif yang terlewatkan oleh model.

4. Nilai Precision 1 menunjukkan bahwa semua prediksi positif yang dibuat oleh model adalah benar. Tidak ada satupun sampel negatif yang salah diklasifikasikan sebagai positif.

5. Nilai F1 Score 1 menunjukkan keseimbangan yang optimal antara Recall dan Precision. Model memiliki tingkat Recall yang tinggi dan tingkat Precision yang tinggi pula.

6. Nilai Kappa 1 menunjukkan kesesuaian sempurna antara prediksi model dengan kelas aktual yang terobservasi. Tidak ada perbedaan antara prediksi model dan kelas aktual.

7. Nilai MCC 1 menunjukkan tingkat kesesuaian sempurna antara prediksi model dengan kelas aktual. Koefisien korelasi Matthews mencakup keseluruhan matriks konfusi dan memberikan nilai yang tinggi ketika prediksi dan observasi sama persis.

Hal ini menunjukkan bahwa model memiliki kinerja yang sangat baik dan mampu melakukan klasifikasi dengan akurasi dan keakuratan yang sempurna.
